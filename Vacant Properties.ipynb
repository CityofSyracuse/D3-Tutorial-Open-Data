```python
# We are going to import the necessary Python Libraries
import pandas as pd

# First we are going to download the Vacant Properties dataset and then load the CSV into the Notebook
fpath = "C:/Users/jscharf/Downloads/Vacant_Properties.csv"

df = pd.read_csv(fpath)
df.head()
```

```python
# Group by OwnerName and count the number of properties for each owner
# df['owner'].value_counts() counts the occurrences of each unique owner in the 'owner' column.
# .reset_index() converts the resulting Series into a DataFrame and resets the index, so the counts are now a new column in the DataFrame.
owner_counts_df = df['Owner'].value_counts().reset_index()
owner_counts_df.head
```

```python
# Rename the columns to 'owner' and 'num_properties'
owner_counts_df.columns = ['Owner', 'Property_Count']
owner_counts_df.head()
```

```python
# Save to a CSV
owner_counts_df.to_csv("owner_count.csv")
```
